```python
# import module
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
%matplotlib inline
uploaded = files.upload()
```

```python
df = pd.read_csv("stock.csv")
df.head(13)
```

```python
df = pd.DataFrame({
    'Stock Name': ['Stock A', 'Stock B', 'Stock C', 'Stock D', 'Stock E'],
    'Symbol': ['A', 'B', 'C', 'D', 'E'],
    'Purchase Price': [100, 150, 200, 250, 300],
    'Current Price': [110, 160, 180, 220, 320],
    'Gain/Loss': [10, 10, -20, -30, 20],
    'Dividen/Share': [5, 3, 7, 8, 2],
    'Market Value': [1000, 1500, 500, 2000, 1000],
    'Annual Yield': [0.05, 0.02, 0.04, 0.06, 0.01]
})
```

```python
# a. Histogram - Distribusi data pada atribut Purchase Price, Current Price, dan Gain/Loss
plt.figure(figsize=(10, 6))
plt.hist(df['Purchase Price'], bins=10, alpha=0.5, color='blue', label='Purchase Price')
plt.hist(df['Current Price'], bins=10, alpha=0.5, color='red', label='Current Price')
plt.hist(df['Gain/Loss'], bins=10, alpha=0.5, color='green', label='Gain/Loss')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram - Purchase Price, Current Price, and Gain/Loss')
plt.legend()
plt.show()
```

```python
# b. Atribut Stock Name dan Symbol dengan Dividen / Share 5 tertinggi
top_dividend_stocks = df.nlargest(5, 'Dividen/Share')[['Stock Name', 'Symbol']]
print('Top 5 Stocks by Dividen/Share:')
print(top_dividend_stocks)
```

```python
# c. Atribut Stock Name dan Symbol dengan Market Value 5 tertinggi
top_market_value_stocks = df.nlargest(5, 'Market Value')[['Stock Name', 'Symbol']]
print('Top 5 Stocks by Market Value:')
print(top_market_value_stocks)
```

```python
# d. Atribut Stock Name dan Symbol dengan Annual Yield 7 tertinggi
top_yield_stocks = df.nlargest(7, 'Annual Yield')[['Stock Name', 'Symbol']]
print('Top 7 Stocks by Annual Yield:')
print(top_yield_stocks)
```
